```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Admission_Predict_Ver1.1.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop(columns = ['Serial No.'], inplace = True)
```

```python
df.head()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=1)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(7,activation = 'relu',input_dim = 7))   # for regression type problems output activation function should be linear
model.add(Dense(7,activation = 'relu'))  # adding hidden layer to improve score
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(loss = 'mean_squared_error',optimizer= keras.optimizers.Adam(learning_rate=0.001))
```

```python
history = model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.2)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
